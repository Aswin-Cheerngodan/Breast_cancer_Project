Feature selection

```python
from sklearn.feature_selection import mutual_info_classif

X = cancer_data_transformed
y = cancer_data['Target']

mi_scores = mutual_info_classif(X,y,random_state=42)

mi_df = pd.DataFrame({'features':X.columns.to_list(),'mi_score':mi_scores})
mi_df.sort_values(by='mi_score',ascending=False)
```

we can remove the features that has not mi above 0.05

```python
selected_features = mi_df[mi_df['mi_score']>0.05]['features'].to_list()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


X = cancer_data_transformed[selected_features]
y= cancer_data['Target']

X_train,X_test,y_train,y_test = train_test_split(X,y)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```
